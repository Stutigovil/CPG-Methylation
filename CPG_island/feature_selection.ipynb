```python
import pandas as pd
import numpy as np
import glob
import os

from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_classif
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel
```

```python
DATA_FOLDER = r"C:\Users\Stuti\Desktop\Projects\CGP\sample_batches"   # CHANGE THIS
label_col = "parkinsons_disease"             # your target
```

```python
files = sorted(glob.glob(os.path.join(DATA_FOLDER, "*.csv")))
print("Total files:", len(files))
```

```python
batch_size = 10
chunks = []

for start in range(0, 50, batch_size):
    end = min(start + batch_size, 50)
    batch_files = files[start:end]
    
    print(f"Loading files {start} to {end} ...")
    
    # load batch
    batch_dfs = [pd.read_csv(f) for f in batch_files]
    batch_df = pd.concat(batch_dfs, ignore_index=True)
    
    chunks.append(batch_df)

# Concatenate everything
df = pd.concat(chunks, ignore_index=True)

print("Final combined shape:", df.shape)
```

```python
X = df.drop(columns=[label_col]).select_dtypes("number")
y = df[label_col]

print("X shape:", X.shape)
```

```python
vt = VarianceThreshold(threshold=0.0)
X1 = vt.fit_transform(X)

cols1 = X.columns[vt.get_support()]
print("After Variance Threshold:", X1.shape)
```

```python
k = 10000      # choose top 3000 CpGs

skb = SelectKBest(f_classif, k=k)
X2 = skb.fit_transform(X1, y)

cols2 = cols1[skb.get_support()]
print("After ANOVA:", X2.shape)
```

```python
et = ExtraTreesClassifier(
    n_estimators=1000,
    random_state=42,
    n_jobs=-1
)

et.fit(X2, y)

selector = SelectFromModel(et, prefit=True, threshold="median")

X3 = selector.transform(X2)
cols3 = cols2[selector.get_support()]

print("After ExtraTrees:", X3.shape)
print("Final selected features:", len(cols3))
```

```python
final_features_df = pd.DataFrame(cols3, columns=["CpG"])
final_features_df.to_csv("final_selected_cpgs.csv", index=False)

print("Saved final selected CpGs!")
```
