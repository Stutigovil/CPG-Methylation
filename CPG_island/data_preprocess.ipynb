```python
import numpy as np

X = df.drop(columns=["parkinsons_disease"])
y = df["parkinsons_disease"]

# Keep only numeric CpG columns
X = X.select_dtypes(include=[np.number])
```

```python
group0 = X[y == 0]   # Control
group1 = X[y == 1]   # Parkinson's
```

```python
from scipy.stats import ttest_ind

t_stats = []
p_values = []

for col in X.columns:
    t, p = ttest_ind(group0[col], group1[col], nan_policy="omit")
    t_stats.append(t)
    p_values.append(p)
```

```python
import pandas as pd

results = pd.DataFrame({
    "CpG_Site": X.columns,
    "t_stat": t_stats,
    "p_value": p_values
})

results = results.sort_values("p_value")
results.head(20)
```

```python
significant_cpgs = results[results["p_value"] < 0.05]
significant_cpgs.to_csv("significant_cpgs_ttest.csv", index=False)
```

```python
X_num = df.drop(columns=["parkinsons_disease"]).select_dtypes(include="number")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

# Try cluster sizes from 1 to 10
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_num)
    wcss.append(kmeans.inertia_)

# Plot elbow curve
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS (Within Cluster Sum of Squares)")
plt.title("Elbow Method")
plt.show()
```

```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_classif
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel

# ---------------------------------
# Load data
# ---------------------------------
# X = df_malat1.drop('Category', axis=1)
X = df.drop('Category', axis=1)
y = df['Category'].map({'T2D': 1, 'ND': 0})

# y = df_malat1['Category'].map({'T2D': 1, 'ND': 0})

print("Initial shape:", X.shape)

# ---------------------------------
# 1) Remove constant features
# ---------------------------------
vt = VarianceThreshold(threshold=0.0)
X1 = vt.fit_transform(X)
cols1 = X.columns[vt.get_support()]
print("After VarianceThreshold:", X1.shape)

# ---------------------------------
# 2) Univariate Filtering (ANOVA)
# Keep top 3000 best features
# ---------------------------------
k = 3000
skb = SelectKBest(score_func=f_classif, k=k)
X2 = skb.fit_transform(X1, y)
cols2 = cols1[skb.get_support()]
print("After SelectKBest:", X2.shape)

# ---------------------------------
# 3) Tree-Based Feature Selection
# Keep top ~50% of remaining
# ---------------------------------
et = ExtraTreesClassifier(
    n_estimators=400,
    random_state=42,
    n_jobs=-1
)
et.fit(X2, y)

selector = SelectFromModel(et, prefit=True, threshold="median")
X3 = selector.transform(X2)
cols3 = cols2[selector.get_support()]

print("After ExtraTrees SelectFromModel:", X3.shape)
print("Total selected features:", len(cols3))
print(cols3)
```
